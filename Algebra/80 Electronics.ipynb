```python
x = list()
y = list()
for i in range ( 5,101):
    x.append(i)
for i in range (5, 101):
    equation = (((10770) / (pow(i,2))) - 0.37)
    y.append(equation)
print(x + y)
print("If x = 85.5 | y-value = ",(((10770) / (pow(85.5,2))) - 0.37) )
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.xlabel ( "Diameter in mils" )
plt.ylabel ( " Resistance in ohms")
plt.title ( "Electronics")
plt.show()
```

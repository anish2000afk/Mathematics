```python
import matplotlib.pyplot as plt
x = list()
y = list()
for i in range (20,101):
    x.append(i)
for i in range (20,101):
    equation = ((-.00025 * pow(i,2)) + (0.574 * i) + 44.25)
    y.append(equation)
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.xlabel('Time in years')
plt.ylabel('Life expectancy')
plt.title('Population Statistics')
plt.show()
```

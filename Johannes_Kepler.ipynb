```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0.387,0.723,1.000,1.524,5.203]
y = [0.241,0.615,1.000,1.881,11.860]
plt.plot(x,y)
plt.show()
```

```python
newX = list()
newY = list()
for i in x:
    i = i * pow(i, 0.5)
    newX.append(i)
for i in y:
    i = i * pow(i,0.333333)
    newY.append(i)
plt.plot(newX, newY,x,y)
plt.show()
```
